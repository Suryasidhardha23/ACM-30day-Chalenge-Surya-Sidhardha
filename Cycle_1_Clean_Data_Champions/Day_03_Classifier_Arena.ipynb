```python
import pandas as ps
```

```python
df = ps.read_csv("mental_health_workplace_survey.csv")
```

```python
df.head(6)
```

```python
df['work hoursperday']=df['WorkHoursPerWeek']/7
```

```python
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X = df[['work hoursperday', 'BurnoutLevel']]
y = df['BurnoutRisk']
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.35, random_state=5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score, roc_auc_score, classification_report, confusion_matrix
model = LogisticRegression()
model.fit(X_train, y_train)
prediction = model.predict(X_test)
probs = model.predict_proba(X_test)[:, 1]
log_acc = accuracy_score(y_test, prediction)
log_auc = roc_auc_score(y_test, probs)
log_conf = confusion_matrix(y_test, prediction)
log_report = classification_report(y_test, prediction, output_dict=True)
log_summary_df = ps.DataFrame({'Metric': ['Accuracy', 'ROC-AUC'], 'Logistic Regression': [log_acc, log_auc]})
log_conf_df = ps.DataFrame(log_conf, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
log_report_df = ps.DataFrame(log_report).transpose()
log_summary_df
```

```python
log_conf_df
```

```python
log_report_df
```

```python

model2 = LinearDiscriminantAnalysis()
model2.fit(X_train, y_train)
preds = model2.predict(X_test)
probs = model2.predict_proba(X_test)[:, 1]
lda_acc = accuracy_score(y_test, preds)
lda_auc = roc_auc_score(y_test, probs)
lda_conf = confusion_matrix(y_test, preds)
lda_report = classification_report(y_test, preds, output_dict=True)
lda_summary_df = ps.DataFrame({'Metric': ['Accuracy', 'ROC-AUC'], 'LDA': [lda_acc, lda_auc]})
lda_conf_df = ps.DataFrame(lda_conf, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
lda_report_df = ps.DataFrame(lda_report).transpose()
lda_summary_df
```
