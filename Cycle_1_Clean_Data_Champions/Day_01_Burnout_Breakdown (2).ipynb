```python
import pandas as pd
```

```python
df = pd.read_csv('mental_health_workplace_survey.csv')
```

```python
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].mean())
```

```python
print(df.isnull().sum())
```

```python
df = df[df['WorkHoursPerWeek'] <= 90]
```

```python
import seaborn as sns
sns.boxplot(data=df, x='StressLevel', y='SleepHours')
```

```python
sns.boxplot(data=df, x='MentalHealthDaysOff', y='BurnoutLevel')
```
